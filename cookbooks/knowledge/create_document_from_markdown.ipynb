# Create a document from markdown

## Introduction

This cookbook shows how to create a document from markdown content.

```python
# Jinko specifics imports & initialization
# Please fold this section and do not change
import jinko_helpers as jinko

# Connect to Jinko (see README.md for more options)
jinko.initialize()
```

```python
# Cookbook specifics imports
import os

markdown = """
## Data Import and Model Calibration: hands-on practice

*You will now apply the concepts introduced during this session in a practical exercise, following a workflow similar to what you would use in a real project.*

*Please complete the tasks below in the proposed order.*

**IMPORTANT: please perform the tasks below from the folder associated with your name in your project.** 

## Goal of this hands-on

Calibrate the BCG model on jinkō based on data extracted from literature.

- Open the model you want to calibrate (the non calibrated BCG model is available in your folder)
- Click on **Solve Model → Calibrate**
"""

response = jinko.make_request('/app/v1/document/markdown', method='POST',
    data=markdown,
    options={
        'input_format': 'text/markdown',
        'name': 'My Document',
    }
)
# Get the URL of the resource
print(f"Resource link: {jinko.get_project_item_url_from_response(response)}")
```
