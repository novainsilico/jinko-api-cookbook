# Export a document to LaTeX

## Introduction

This cookbook shows how to export a document to LaTeX and save the resulting ZIP archive to disk.

```python
# Jinko specifics imports & initialization
# Please fold this section and do not change
import jinko_helpers as jinko

# Connect to Jinko (see README.md for more options)
jinko.initialize()
```

```python
# Cookbook specifics imports
import os

# @param {"name":"documentId", "type": "string"}
# Fill the short Id of your Document (ex: do-uNcP-kEA1)
document_id = "do-uNcP-kEA1"

output_file=f"{document_id}.zip"
jinko.make_request(
  path=f"/app/v1/document/{document_id}/export/latex",
  output_file=output_file
)

print(f"Document successfully exported to {output_file}")
```
