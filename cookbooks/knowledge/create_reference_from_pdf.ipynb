# Create a reference

## Introduction

This cookbook shows how to create a reference from a PDF file.

```python
# Jinko specifics imports & initialization
# Please fold this section and do not change
import jinko_helpers as jinko

# Connect to Jinko (see README.md for more options)
jinko.initialize()
```

```python
# Cookbook specifics imports
import os

resources_dir = os.path.normpath("./resources")
file_path = os.path.join(resources_dir, "paper.pdf")

with open(file_path, 'rb') as file:
    response = jinko.make_request('/app/v1/reference/file', method='POST',
        file=file,
    )
    # Get the URL of the resource
    print(f"Resource link: {jinko.get_project_item_url_from_response(response)}")
```
